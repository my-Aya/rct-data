```python
import os
import pandas as pd
from datetime import datetime
import numpy as np

# Change to the folder containing all the demographics CSV files you downloaded
os.chdir('/Users/garvert/Documents/Alena/RCT/rct-data')

files = [x for x in os.listdir() if x.find('.xlsx') >= 0]

total_dur = []
ids = []
for file in files:
    df = pd.read_excel(file)

    df_new = df[['uid', 'Start Date (UTC)', 'Submit Date (UTC)']]

    df_new['Start Date (UTC)'] = pd.to_datetime(df_new['Start Date (UTC)'])
    df_new['Submit Date (UTC)'] = pd.to_datetime(df_new['Submit Date (UTC)'])

    # Compute duration and convert to seconds
    diff = (df_new['Submit Date (UTC)'] - df_new['Start Date (UTC)'])
    df_new['duration (sec)'] = (diff.dt.days * 24 * 60 * 60 ) + (diff.dt.seconds)

    df_new.to_csv(file[0:4] + '_duration.csv', sep='\t')

    total_dur.append(np.median(df_new['duration (sec)']))

print(total_dur)

```
